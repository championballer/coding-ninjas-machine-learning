### Handling Strings

```python
df["Gender"] = "Female"
df.head()
```

```python
df.iloc[0:10,6] = "Male"
```

```python
df.head()
```

```python
df["Gender"].value_counts()
```

```python
def male_ohe(s):
    if s == "Male":
        return 1
    else:
        return 0
```

```python
def female_ohe(s):
    if s == "Female":
        return 1
    else:
        return 0
```

```python
df["Male"] = df["Gender"].apply(male_ohe)
df.head()
```

```python
df["FeMale"] = df["Gender"].apply(female_ohe)
```
